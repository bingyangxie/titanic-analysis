Task #03：鐵達尼號存活預測專題實作（2） - 建立第一個 Baseline 版本

```python
import pandas as pd
import numpy as np
url_train = 'https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv'
df_train = pd.read_csv(url_train)
df_train.info
```

```python
# 查詢各欄位的缺失值個數
missing_values_count = df_train.isnull().sum()

print(missing_values_count)
```

```python
# 刪除'Cabin'欄位
df_train.drop('Cabin', axis=1, inplace=True)
# 刪除'Name'欄位
df_train.drop('Name', axis=1, inplace=True)
# 刪除'Ticket'欄位
df_train.drop('Ticket', axis=1, inplace=True)
# 計算'Age'欄位的眾數
age_mode = df_train['Age'].mode()[0]
# 用眾數填補'Age'欄位的缺失值
df_train['Age'].fillna(age_mode, inplace=True)
# 使用Label編碼將'Sex'和'Embarked'欄位轉換成數值型
df_train['Sex'] = df_train['Sex'].map({'female': 0, 'male': 1})
df_train['Embarked'] = df_train['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
# 計算'Embarked'欄位的眾數
embarked_mode = df_train['Embarked'].mode()[0]
# 用眾數填補'Embarked'欄位的缺失值
df_train['Embarked'].fillna(embarked_mode, inplace=True)
df_train
```

```python
# 查詢各欄位的缺失值個數
missing_values_count = df_train.isnull().sum()

print(missing_values_count)
```

```python
columns_X = set(df_train.columns) - {'Survived'}
columns_y = ['Survived']


train_X = df_train[columns_X]
train_y = df_train[columns_y]


from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score


log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(),cv=5,scoring='accuracy')
print(scores)
```
