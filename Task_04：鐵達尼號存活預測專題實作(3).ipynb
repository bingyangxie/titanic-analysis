```python
import pandas as pd


url_train = 'https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv'
df_train = pd.read_csv(url_train)

df_sex = pd.get_dummies(df_train['Sex'])
df_embarked = pd.get_dummies(df_train['Embarked'])
# Display the original DataFrame with OneHotEncoding of 'Sex'
df_train = pd.concat([df_train, df_sex, df_embarked], axis=1)
# Drop the original 'Sex' and 'Embarked' columns
df_train.drop(['Sex', 'Embarked'], axis=1, inplace=True)
df_train
```

```python
# 刪除'Cabin'欄位
df_train.drop('Cabin', axis=1, inplace=True)
# 刪除'Name'欄位
df_train.drop('Name', axis=1, inplace=True)
# 刪除'Ticket'欄位
df_train.drop('Ticket', axis=1, inplace=True)
# 計算'Age'欄位的眾數
age_mode = df_train['Age'].mode()[0]
# 用眾數填補'Age'欄位的缺失值
df_train['Age'].fillna(age_mode, inplace=True)
```

```python
df_train
```

```python
columns_X = set(df_train.columns) - {'Survived'}
columns_y = ['Survived']


train_X = df_train[columns_X]
train_y = df_train[columns_y]


from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score


log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(),cv=5,scoring='accuracy')
print(scores)
```
